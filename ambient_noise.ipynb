```python
import pandas as pd
import matplotlib.pyplot as plt

# Folder path containing all .ods files for each year
folder_path = "/mnt/windows/my codes/python/noise research/env research/"
# List of years for each file (assuming filenames include the year)
years = ["2011-2012", "2012-2013", "2013-2014","2014-2015","2019-2020", "2020-2021", "2022-2023" ]  # Add more years as needed
# Read the .ods file
for year in years:
    ods_file_path = f"{folder_path}/{year}.ods"
    df = pd.read_excel(ods_file_path, engine='odf')

    # Forward fill the City column to fill NaN values with the previous city name
    df["City"] = df["City"].ffill()

    # Convert 'Day Time db(A)' and 'Night Time db(A)' to numeric, forcing errors to NaN
    df["Day Time db(A)"] = pd.to_numeric(df["Day Time db(A)"], errors='coerce')
    df["Night Time db(A)"] = pd.to_numeric(df["Night Time db(A)"], errors='coerce')

    # Check for missing values after conversion
    print("Missing values in the DataFrame:")
    print(df.isnull().sum())

    # Group by city and analyze noise levels
    city_analysis = {}

    for city, group in df.groupby("City"):
        # Calculate statistics
        avg_day_time = group["Day Time db(A)"].mean()
        avg_night_time = group["Night Time db(A)"].mean()
        max_day_time = group["Day Time db(A)"].max()
        max_night_time = group["Night Time db(A)"].max()
        min_day_time = group["Day Time db(A)"].min()
        min_night_time = group["Night Time db(A)"].min()
        
        # Store the results in a dictionary
        city_analysis[city] = {
            'Average Day Time (dB)': avg_day_time,
            'Average Night Time (dB)': avg_night_time,
            'Max Day Time (dB)': max_day_time,
            'Max Night Time (dB)': max_night_time,
            'Min Day Time (dB)': min_day_time,
            'Min Night Time (dB)': min_night_time,
            'Count of Stations': group.shape[0]
        }
        
        # Print the analysis for each city
        print(f"City: {city}")
        print(f"Average Day Time (dB): {avg_day_time:.2f}")
        print(f"Average Night Time (dB): {avg_night_time:.2f}")
        print(f"Max Day Time (dB): {max_day_time:.2f}")
        print(f"Max Night Time (dB): {max_night_time:.2f}")
        print(f"Min Day Time (dB): {min_day_time:.2f}")
        print(f"Min Night Time (dB): {min_night_time:.2f}")
        print(f"Count of Stations: {group.shape[0]}")
        print()

    # Optional: Create a bar plot to compare average noise levels between cities
    cities = list(city_analysis.keys())
    avg_day_levels = [city_analysis[city]['Average Day Time (dB)'] for city in cities]
    avg_night_levels = [city_analysis[city]['Average Night Time (dB)'] for city in cities]

    x = range(len(cities))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    bars1 = ax.bar(x, avg_day_levels, width, label='Avg Day Time (dB)', color='orange')
    bars2 = ax.bar([p + width for p in x], avg_night_levels, width, label='Avg Night Time (dB)', color='blue')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Noise Level (dB)')
    ax.set_title('Average Noise Levels by City')
    ax.set_xticks([p + width / 2 for p in x])
    ax.set_xticklabels(cities)
    ax.legend()

    # Show the plot
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


```
